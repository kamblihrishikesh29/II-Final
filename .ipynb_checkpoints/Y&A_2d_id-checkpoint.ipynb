## FOR 2-D Identity Speckle Pattern

```python
# FOR 2-D Identity
scene = np.loadtxt('C:/Users/kambl/Johns Hopkins/Siya Zhang - Lab data/FinalProj/scene_2d_59.csv', delimiter=',')
bs1 = np.loadtxt('C:/Users/kambl/Johns Hopkins/Siya Zhang - Lab data/FinalProj/blackimg_59.csv', delimiter=',')
bs2 = np.loadtxt('C:/Users/kambl/Johns Hopkins/Siya Zhang - Lab data/FinalProj/blackimg2_59.csv', delimiter=',')
masks_2d_id = np.loadtxt('C:/Users/kambl/Johns Hopkins/Siya Zhang - Lab data/FinalProj/mask_2d_59.csv', delimiter=',')
```

```python
bs = (bs1 + bs2)/2
bs.shape
```

```python
a1 = scene.reshape(1024,1280,256)
for i in range(256):
    a1[:,:,i] = a1[:,:,i] - bs
    a1[:,:,i] = np.where(a1[:,:,i]<0, 0, a1[:,:,i])
```

```python
masks_2d_id.shape
```

```python
A_2d_id = np.copy(masks_2d_id).reshape(256,256)
```

```python
plt.imshow(A_2d_id, cmap='gray')
```

```python
# 16x16 pixels square in scene with left top(100,260) to bottom right()
Y_2d_id = np.zeros([256,1])
for i in range(256):
    Y_2d_id[i,0] = np.sum(a1[100:845,260:1018,i])
```

```python
a2 = Y_2d_id.reshape(16,16)
plt.imshow(a2, cmap='gray')
```

```python
ab = np.linalg.pinv(A_2d_id, rcond=1e-1)
xhat = np.matmul(ab,Y_2d_id)
```

```python
plt.imshow(xhat.reshape(16,16), cmap='gray')
```
